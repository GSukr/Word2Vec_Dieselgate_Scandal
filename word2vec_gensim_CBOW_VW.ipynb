```python
import pandas as pd # for reading the file
import gensim # for word2vec model
import logging # for logs
import re
```

```python
from sqlalchemy import create_engine
disk_engine = create_engine('sqlite:///all_tweets.db')
df = pd.read_sql('SELECT * FROM tweets_table_new', disk_engine)
```

```python
df.head(2)
```

```python
sentences = df.cleaned_text
```

```python
def clean_sentences(sentences, remove_stopwords=False):
    clean = []
    for sentence in sentences:
        URL = "(?:https?:\s?\\/\\/\s?(?:www\\.|(?!www))[^\\s\\.]+\\.[^\\s]{2,}|www\\.[^\\s]+\\.[^\\s]{2,})"
        URL_pattern = re.compile(URL)
        sentence = re.sub(URL_pattern, '', str(sentence))#Remove URL
        sentence = re.sub("[^a-zA-Z]"," ", str(sentence)) # Remove non-letters
        sentence = sentence.split() # Convert words to lower case
        clean.append(sentence)
    return clean
```

```python
sentences_clean = clean_sentences(sentences) # Make the list of sentences representing list of words
```

```python
print (sentences_clean[12021])
```

```python
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s',\
    level=logging.INFO)
```

```python
num_features = 500    # Word vector dimensionality                      
min_word_count = 10   # Minimum word count                        
num_workers = 4       # Number of threads to run in parallel
context = 3         # Context window size                                                                                    
```

```python
from gensim.models import word2vec
print ("Training model...")
model = word2vec.Word2Vec(sentences_clean, workers=num_workers, \
            size=num_features, min_count = min_word_count, \
            window = context, iter = 5)
```

```python
model.init_sims(replace=True)
```

```python
model_name = "500features_10minwords_3context_tweets" # to save the model
model.save(model_name)
```

```python
model.wv.most_similar("manipulation", topn = 30) #to find the similar words
```

```python
model.most_similar(positive=["scandal"],negative = ["protection"],topn=10)
```
